```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_wine
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = load_wine()
data
data.data.shape
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(data.data, data.target, test_size=0.3,random_state=0)
```

```python
clf = DecisionTreeClassifier(random_state=0)
clf = clf.fit(Xtrain, Ytrain)
clf.score(Xtest, Ytest)

```

```python
rfc = RandomForestClassifier(random_state=0)
rfc = rfc.fit(Xtrain, Ytrain)
rfc.score(Xtest, Ytest)
```

```python
from sklearn.model_selection import cross_val_score

clf1 = DecisionTreeClassifier(random_state=0)
rfc1 = RandomForestClassifier(random_state=0)

clf_score= cross_val_score(clf1, data.data, data.target, cv=10)
rfc_score= cross_val_score(rfc1, data.data, data.target, cv=10)
print(clf_score.mean(), rfc_score.mean())
```

```python
clf_s = []
rfc_s = []

for i in range(1,11):
    clf = DecisionTreeClassifier()
    rfc = RandomForestClassifier()
    clf_score= cross_val_score(clf, data.data, data.target, cv=10).mean()
    clf_s.append(clf_score)
    rfc_score= cross_val_score(rfc, data.data, data.target, cv=10).mean()
    rfc_s.append(rfc_score)

plt.plot(range(1,11),clf_s,color="red",label="Decision")
plt.plot(range(1,11),rfc_s,color="blue",label="RandomForest")  
plt.legend()
plt.show()
```

```python
# test for n_estimators

superpa = []
for i in range(1,201):
    rfc = RandomForestClassifier(n_estimators=i,n_jobs=-1)
    rfc_score = cross_val_score(rfc,data.data,data.target,cv=10).mean()
    superpa.append(rfc_score)
```

```python
print(max(superpa),superpa.index(max(superpa)))
plt.figure(figsize=[20,5])
plt.plot(range(1,201),superpa)
plt.show()

```
