```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd
```

```python
data = load_boston()
rfc = RandomForestRegressor(random_state=100)
cross_val_score(rfc, data.data, data.target, cv=10,scoring="neg_mean_squared_error")
```

```python
# here is the typical code to show the score of the model
# if you dont know which scoring method to use, you can use the following code to check
import sklearn
sorted(sklearn.metrics.SCORERS.keys())
```

```python
paremeters = {"n_estimators": [100,200,300]
            #  ,"creterion": ["mse","mae"]
              ,"max_depth": np.arange(1,10,1)
              ,"min_samples_split": np.arange(2,10,1)
              ,"min_samples_leaf": np.arange(1,10,1)
              ,"min_impurity_decrease": np.arange(0,1,0.2)}


rfc = RandomForestRegressor(random_state=100)

from sklearn.model_selection import GridSearchCV
GS = GridSearchCV(rfc,paremeters,cv=10)
GS.fit(data.data,data.target)


```

```python
print(GS.best_params_)
print(GS.best_score_)
print(GS.best_estimator_)

```
